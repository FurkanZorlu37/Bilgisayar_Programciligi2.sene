```python
import sqlite3

conn = sqlite3.connect(':memory:')
cursor = conn.cursor()

cursor.execute('''
    CREATE TABLE IF NOT EXISTS users
    (id INTEGER PRIMARY KEY,
    name TEXT NOT NULL,
    soyad TEXT NOT NULL,
    classes TEXT NOT NULL,
    notes1 INTEGER NOT NULL,
    notes2 INTEGER NOT NULL,
    ortalama BOOLEAN NOT NULL,
    ortalamagibi TEXT NOT NULL
    )
''')

def add_user_data(cursor, conn):
    ad = str(input("Adı girin: "));
    soyad=str(input("Soyadı girin: "));
    classes = str(input("Sınıfı girin: "));
    notes1=int(input("Not 1'i girin: "));
    notes2=int(input("Not 2'yi girin: "));

    ortalama_value = notes1 * 0.40 + notes2 * 0.60;
    basarilibasarisiz = "";
    if ortalama_value >= 55:
        ortalama_bool = True;
        basarilibasarisiz = "basarılı !"
    else:
        ortalama_bool = False;
        basarilibasarisiz = "basarısız !"

    try:
        cursor.execute("INSERT INTO users (name, soyad, classes, notes1, notes2, ortalama, ortalamagibi) VALUES (?, ?, ?, ?, ?, ?, ?)",
         (ad, soyad, classes, notes1, notes2, ortalama_bool, basarilibasarisiz))
        conn.commit()
        print("Kullanıcı verisi başarıyla eklendi.")
    except sqlite3.IntegrityError as e:
        print(f"Hata: Veri eklenirken bir bütünlük hatası oluştu: {e}")
        conn.rollback()
    except Exception as e:
        print(f"Hata: Beklenmedik bir hata oluştu: {e}")
        conn.rollback()


while True:
    add_user_data(cursor, conn)
    devam_et = input("Başka kullanıcı eklemek ister misiniz? (evet/hayır): ").lower()
    if devam_et != 'evet':
        break

print("-----------------------------------")
cursor.execute("SELECT * FROM users")
rows = cursor.fetchall()
print("Veritabanındaki Kullanıcılar: ")
for row in rows:
    print(row)
conn.close()
```
