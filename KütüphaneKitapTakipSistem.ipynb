```python
import sqlite3;
import datetime;

def kitap_ekle(kitap_adi, yazar):
    conn = sqlite3.connect('database.db');
    cursor = conn.cursor();
    # sql I yememek icin sorguları direk larak ? ? kulanılıyor //parametreli sorgu kullanılıyor
    cursor.execute("INSERT INTO kitaplar (kitap_adi, yazar) VALUES (?, ?)", (kitap_adi, yazar));
    conn.commit();
    conn.close();

def kitaplari_listele():
    conn = sqlite3.connect('database.db');
    cursor = conn.cursor();
    # Bu sorguda kullanıcıdan veri alınmadığı için SQL Injection riski yoktur.
    cursor.execute("SELECT * FROM kitaplar");
    kitaplar = cursor.fetchall();
    conn.close();
    if kitaplar:
        print("\n--- Kitap Listesi ---")
        for kitap in kitaplar:
            print(f"ID: {kitap[0]}, Kitap Adı: {kitap[1]}, Yazar: {kitap[2]}, Ekleme Tarihi: {kitap[3]}")
        print("---------------------\n")
    else:
        print("Veritabanında hiç kitap bulunmuyor.\n")

def kitap_tarih_hesapla():
  conn = sqlite3.connect('database.db');
  cursor = conn.cursor();
  # Bu sorguda kullanıcıdan veri alınmadığı için SQL Injection riski yoktur.
  cursor.execute("SELECT kitap_adi, ekleme_tarihi FROM kitaplar");
  kitap_tarihleri = cursor.fetchall();
  conn.close();

  if kitap_tarihleri:
      print("\n--- Kitap Kayıt Süreleri ---")
      now = datetime.datetime.now();
      for kitap_adi, ekleme_tarihi_str in kitap_tarihleri:
          # 'YYYY-MM-DD HH:MM:SS' yıl-ay-gün ve saat dakika ve saniye olucak sekilde gorunume alıyor.
          ekleme_tarihi = datetime.datetime.strptime(ekleme_tarihi_str, '%Y-%m-%d %H:%M:%S')
          gecen_sure = now - ekleme_tarihi

          # Formatını daha duzenli hale getirmeye ugrasıyorum.
          parts = []
          if gecen_sure.days > 0:
              parts.append(f"{gecen_sure.days} gün")
          seconds = gecen_sure.seconds
          hours = seconds // 3600
          minutes = (seconds % 3600) // 60
          if hours > 0:
              parts.append(f"{hours} saat")
          if minutes > 0:
              parts.append(f"{minutes} dakika")

          formatted_sure = " ".join(parts) if parts else "birkaç saniye"

          print(f"'{kitap_adi}' adlı kitap {formatted_sure} önce eklendi.")
      print("----------------------------\n")
  else:
      print("Veritabanında kayıtlı kitap bulunmuyor.\n")

def kitap_sil(kitap_id):
    conn = sqlite3.connect('database.db');
    cursor = conn.cursor();
    # SQL-i önlemek için parametreli sorgu kullanılıyor
    cursor.execute("DELETE FROM kitaplar WHERE id = ?", (kitap_id,));
    conn.commit();
    conn.close();
    if cursor.rowcount > 0:
        print(f"{kitap_id} ID'li kitap başarıyla silindi.\n")
    else:
        print(f"{kitap_id} ID'li kitap bulunamadı veya silinemedi.\n")

now = datetime.datetime.now();

conn = sqlite3.connect('database.db');
print("Opened database successfully");
cursor = conn.cursor();

cursor.execute('''
CREATE TABLE IF NOT EXISTS kitaplar (
    id INTEGER PRIMARY KEY AUTOINCREMENT,
    kitap_adi TEXT NOT NULL,
    yazar TEXT NOT NULL,
    ekleme_tarihi TEXT DEFAULT CURRENT_TIMESTAMP
);
''');
conn.commit();
conn.close();
print("Table 'kitaplar' created and connection closed successfully.");

# Sonsuz dongu ve kulanıcı kulanımı:
while True:
    print("---------------------------------")
    print("Yapmak istediğiniz işlemi seçin:")
    print("1: Kitap Ekle")
    print("2: Kitapları Listele")
    print("3: Kitapların Kayıt Süresini Hesapla")
    print("4: Kitap Sil")
    print("0: Çıkış")
    secim = input("Seçiminiz: ")

    if secim == '1':
        kitap_adi = input("Kitap Adı: ")
        yazar = input("Yazar: ")
        kitap_ekle(kitap_adi, yazar)
        print(f"'{kitap_adi}' adlı kitap eklendi.")
    elif secim == '2':
        kitaplari_listele()#Direk functiona aldım ve oradan devam etirdim
    elif secim == '3':
        kitap_tarih_hesapla()#Direk functiona aldım ve oradan devam etirdim
    elif secim == '4':
        kitap_id = input("silmek istediginiz kitabın id'sini girin: ")
        try:
            kitap_sil(int(kitap_id))
        except ValueError:
            print("Geçersiz id. Lütfen sayı girin.")
    elif secim == '0':
        print("Programdan çıkılıyor...")
        break
    else:
        print("Geçersiz seçim. Lütfen tekrar deneyin.")
```
