```python
# Bu fonksiyon, 2000 yılından itibaren belirli bir tarihe kadar geçen toplam gün sayısını hesaplar.
# Artık yılları dikkate alarak doğru hesaplama yapar.
def Totalgunhesabi(hedef_yil, hedef_ay, hedef_gun):

    # Bir yılın artık yıl olup olmadığını kontrol eden yardımcı fonksiyon.
    # Artık yıl: 4'e bölünür, ancak 100'e bölünmez. Veya 400'e bölünür.
    def is_leap(year):
        return (year % 4 == 0 and year % 100 != 0) or (year % 400 == 0)

    # Her ayın gün sayısını tutan liste. İlk eleman (index 0) kullanılmaz.
    days_in_month = [0, 31, 28, 31, 30, 31, 30, 31, 31, 30, 31, 30, 31]
    total_gunler = 0 # Toplam gün sayısını tutacak değişken

    # 2000 yılından hedef yıla kadar olan tüm yılların günlerini toplar.
    # Her yılın artık yıl olup olmamasına göre 365 veya 366 gün ekler.
    for year in range(2000, hedef_yil):
        total_gunler += 366 if is_leap(year) else 365

    # Hedef yıl içindeki ayların günlerini toplar.
    # Eğer ay Şubat ise ve hedef yıl artık yılsa 29 gün ekler.
    for month in range(1, hedef_ay):
        if month == 2 and is_leap(hedef_yil):
            total_gunler += 29
        else:
            total_gunler += days_in_month[month]

    # Son olarak, hedef ayın günlerini ekler.
    total_gunler += hedef_gun

    return total_gunler # Hesaplanan toplam gün sayısını döndürür

# Ayların gün sayılarını tutan liste (yukarıdaki fonksiyonda da kullanılıyor).
days_in_month = [0, 31, 28, 31, 30, 31, 30, 31, 31, 30, 31, 30, 31]

# Kullanıcıdan geçerli bir yıl girmesini isteyen döngü.
# Yıl 2000'den büyük ve 10000'den küçük olmalıdır.
while True:
    yil = int(input("Lütfen 2000 yılından sonrasında bir yıl girin: "))
    if yil <= 2000 or yil > 10000:
        print("Lütfen geçerli bir yıl girin.")
    else:
        break

# Kullanıcıdan geçerli bir ay girmesini isteyen döngü.
# Ay 1 ile 12 arasında olmalıdır.
while True:
    ay = int(input("Lütfen 1 ila 12 arasından seçtiğiniz ayı giriniz: "))
    if ay < 1 or ay > 12:
        print("Lütfen geçerli bir ay girin.")
    else:
      break

# Kullanıcıdan geçerli bir gün girmesini isteyen döngü.
# Girilen gün, seçilen ayın ve yılın (artık yıl kontrolü ile) gün sayısına uygun olmalıdır.
while True:
    gun = int(input("Lütfen 1 ila 31 arası sayı giriniz: "))

    # Seçilen ayın maksimum gün sayısını belirler.
    max_day = days_in_month[ay]
    # Eğer Şubat ayı ise ve yıl artık yıl ise, maksimum gün sayısını 29 olarak ayarlar.
    if ay == 2 and ((yil % 4 == 0 and yil % 100 != 0) or (yil % 400 == 0)):
        max_day = 29

    # Girilen günün geçerli olup olmadığını kontrol eder.
    if gun < 1 or gun > max_day:
        print(f"Lütfen {ay} ayı için geçerli bir gün girin (1 ile {max_day} arası).")
        continue # Geçersizse tekrar gün girmesini ister

    break # Geçerliyse döngüden çıkar

# Hesaplamayı yapmak için Totalgunhesabi fonksiyonunu çağırır.
A = Totalgunhesabi(yil, ay, gun)

# Sonucu ekrana yazdırır.
print(A)
```
