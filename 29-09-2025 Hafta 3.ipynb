```python
print("Ders Notu Hesaplama")
ogrenci_adi = input("Ögrenci adı: ") # Added student name input
ogrenci_no = input("Ögrenci no: ")
Nesne_Vize_Notu = int(input("Nesne Tabanlı Programlama Vize Notu: "))
Nesne_Final_Notu = int(input("Nesne Tabanlı Programlama Final Notu: "))

Gorsel_Vize_Notu = int(input("Görsel Programlama Vize Notu: "))
Gorsel_Final_Notu = int(input("Görsel Programlama Final Notu: "))

Programlama_Temelleri_Vize_Notu = int(input("Programlama Temelleri Vize Notu: "))
Programlama_Temelleri_Final_Notu = int(input("Programlama Temelleri Final Notu: "))

Mobil_Programlama_Vize_Notu = int(input("Mobil Programlama Vize Notu: "))
Mobil_Programlama_Final_Notu = int(input("Mobil Programlama Final Notu: "))

#Ortalama alma - 4 e 6 sekilde genel olarak derslerde kulandıgımız.
Nesne_Ortalama = Nesne_Vize_Notu * 0.4 + Nesne_Final_Notu * 0.6
Gorsel_Ortalama = Gorsel_Vize_Notu * 0.4 + Gorsel_Final_Notu * 0.6
Programlama_Temelleri_Ortalama = Programlama_Temelleri_Vize_Notu * 0.4 + Programlama_Temelleri_Final_Notu * 0.6
Mobil_Programlama_Ortalama = Mobil_Programlama_Vize_Notu * 0.4 + Mobil_Programlama_Final_Notu * 0.6

# direk tek tek yapmak yerine function ici atıp tekrardan yapma
def notu_harfe_cevir(notu):
    if notu >= 90:
        return "AA"
    elif notu >= 85:
        return "BA"
    elif notu >= 80:
        return "BB"
    elif notu >= 75:
        return "CB"
    elif notu >= 70:
        return "CC"
    elif notu >= 65:
        return "DC"
    elif notu >= 60:
        return "DD"
    elif notu >= 50:
        return "FD"
    else:
        return "FF"

# ortalama notları half notuna cevirme
Nesne_Harf_Notu = notu_harfe_cevir(Nesne_Ortalama)
Gorsel_Harf_Notu = notu_harfe_cevir(Gorsel_Ortalama)
Programlama_Temelleri_Harf_Notu = notu_harfe_cevir(Programlama_Temelleri_Ortalama)
Mobil_Programlama_Harf_Notu = notu_harfe_cevir(Mobil_Programlama_Ortalama)

# Formatı duzgun bir sekilde gosterme
print(f"Ögrenci Adı: {ogrenci_adi}")
print(f"Ögrenci Adı: {ogrenci_no}")
print("\nDers Notları:")
print("-" * 60)
print(f"{'Ders Adı':<30}{'Vize':<10}{'Final':<10}{'Ders Sonu':<10}")
print("-" * 60)
print(f"{'Nesne Tabanlı Programlama':<30}{Nesne_Vize_Notu:<10}{Nesne_Final_Notu:<10}{Nesne_Harf_Notu:<10}")
print(f"{'Görsel Programlama':<30}{Gorsel_Vize_Notu:<10}{Gorsel_Final_Notu:<10}{Gorsel_Harf_Notu:<10}")
print(f"{'Programlama Temelleri':<30}{Programlama_Temelleri_Vize_Notu:<10}{Programlama_Temelleri_Final_Notu:<10}{Programlama_Temelleri_Harf_Notu:<10}")
print(f"{'Mobil Programlama':<30}{Mobil_Programlama_Vize_Notu:<10}{Mobil_Programlama_Final_Notu:<10}{Mobil_Programlama_Harf_Notu:<10}")
print("-" * 60)
```
