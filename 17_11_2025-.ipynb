```python
try:
    with open("bilgisayar.txt", "r", encoding="utf-8") as f:
        content = f.read()
        print(content)
except FileNotFoundError:
    print("bilgisayar.txt not found. Creating it with some default content.")
    with open("bilgisayar.txt", "w", encoding="utf-8") as f:
        f.write("Bu bir örnek metindir.\n")
    with open("bilgisayar.txt", "r", encoding="utf-8") as f:
        content = f.read()
        print(content)
```

```python
phone_book = {}

def add_contact(name, phone_number):
    phone_book[name] = phone_number
    print(f"{name}, telefon rehberine eklendi.")

def view_contacts():
    if not phone_book:
        print("Telefon rehberi boş.")
    else:
        print("\nTelefon Rehberi:")
        for name, phone_number in phone_book.items():
            print(f"Ad: {name}, Telefon: {phone_number}")

def save_contacts_to_file(filename="telefon_rehberi.txt"):
    try:
        with open(filename, "w", encoding="utf-8") as f:
            for name, phone_number in phone_book.items():
                f.write(f"Ad: {name}, Telefon: {phone_number}\n")
        print(f"Telefon rehberi '{filename}' dosyasına başarıyla kaydedildi.")
    except Exception as e:
        print(f"Dosyaya kaydederken bir hata oluştu: {e}")

# kaydetme
print("Kişi eklemek için ad ve telefon numarası girin. Çıkmak için 'çıkış' yazın.")
while True:
    ad = input("Adı girin (çıkmak için 'çıkış'): ")
    if ad.lower() == 'çıkış':
        break
    Tel = input(f"{ad} için telefon numarasını girin: ")
    add_contact(ad, Tel)

view_contacts()

# dosyaya kaydetmek.
save_contacts_to_file()
```

```python
with open("telefon_rehberi.txt", "r", encoding="utf-8") as f:
    content = f.read()
    print(content)
```
